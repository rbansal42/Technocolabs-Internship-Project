```python
total_cells = np.product(df.shape)
missing_values_count = df.isnull().sum()
total_missing = missing_values_count.sum()

# percent of data that is missing
percent_missing = (total_missing/total_cells) * 100
print(percent_missing)
```

#Dealing with missing Values


```python
nan_values = df.isna().any()

print (nan_values)
```

```python
perc = 20
min_count =  int(((100-perc)/100)*df.shape[0] + 1)
df2 = df.dropna( axis=1, 
                thresh=min_count)
```

```python
missing_data = pd.DataFrame({'total_missing': df2.isnull().sum(), 'percentage_missing': (df2.isnull().sum()/134529)*100})
missing_data[missing_data['total_missing'] != 0]
```

```python
df3  = df2.dropna()
missing_data = pd.DataFrame({'total_missing': df3.isnull().sum(), 'percentage_missing': (df3.isnull().sum()/134529)*100})
missing_data[missing_data['total_missing'] != 0]
```
